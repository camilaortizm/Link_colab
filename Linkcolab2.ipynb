```python
from abc import ABC, abstractmethod

# 1. Abstracción
class Empleado(ABC):
    def __init__(self, nombre, id_empleado):
        self.nombre = nombre
        self.__id_empleado = id_empleado  # Encapsulamiento

    @abstractmethod
    def calcular_pago(self):
        pass

# 3. Herencia
class EmpleadoTiempoCompleto(Empleado):
    def __init__(self, nombre, id_empleado, salario_mensual):
        super().__init__(nombre, id_empleado)
        self.salario_mensual = salario_mensual

    # 4. Polimorfismo
    def calcular_pago(self):
        return self.salario_mensual

class EmpleadoPorHoras(Empleado):
    def __init__(self, nombre, id_empleado, horas_trabajadas, tarifa_hora):
        super().__init__(nombre, id_empleado)
        self.horas_trabajadas = horas_trabajadas
        self.tarifa_hora = tarifa_hora

    def calcular_pago(self):
        return self.horas_trabajadas * self.tarifa_hora

# Uso
emp1 = EmpleadoTiempoCompleto("Ana", "E001", 3000)
emp2 = EmpleadoPorHoras("Carlos", "E002", 40, 20)

print(f"{emp1.nombre} cobra: {emp1.calcular_pago()}")
print(f"{emp2.nombre} cobra: {emp2.calcular_pago()}")
```

```python
from abc import ABC, abstractmethod

# 1. Abstracción
class Suscripcion(ABC):
    def __init__(self, usuario, precio_base):
        self.usuario = usuario
        self._precio_base = precio_base # Encapsulamiento protegido

    @abstractmethod
    def obtener_beneficios(self):
        pass

    @abstractmethod
    def calcular_precio(self):
        pass

# 3. Herencia
class PlanBasico(Suscripcion):
    def obtener_beneficios(self):
        return "1 pantalla, calidad SD."

    # 4. Polimorfismo: el precio base es el precio final para el plan básico
    def calcular_precio(self):
        return self._precio_base

class PlanPremium(Suscripcion):
    def obtener_beneficios(self):
        # 4. Polimorfismo: el beneficio es distinto
        return "4 pantallas simultáneas, calidad 4K UHD."

    def calcular_precio(self):
        return self._precio_base + 10  # El premium cuesta un extra

# Uso
# Marta tiene 1 pantalla, calidad SD y paga $8.00 (Plan Basico)
user1 = PlanBasico("Marta", 8.00)
# Pedro tiene 4 pantallas simultáneas, calidad 4K UHD y paga $19.00 (Plan Premium)
# Para que el total sea $19.00, el precio_base del PlanPremium debe ser $9.00 (9 + 10 = 19)
user2 = PlanPremium("Pedro", 9.00)

print(f"{user1.usuario}: {user1.obtener_beneficios()} Total: ${user1.calcular_precio()}")
print(f"{user2.usuario}: {user2.obtener_beneficios()} Total: ${user2.calcular_precio()}")
```

```python
from abc import ABC, abstractmethod

# 1. ABSTRACCIÓN: Definimos una clase base 'Medicamento' con métodos abstractos
# que deben ser implementados por sus subclases. Esto asegura que todos los
# medicamentos tengan ciertas funcionalidades mínimas (como `tipo_medicamento`).
class Medicamento(ABC):
    def __init__(self, nombre, dosis_mg, precio):
        self.nombre = nombre
        self.dosis_mg = dosis_mg
        # 2. ENCAPSULAMIENTO: El precio es un atributo privado (`__precio`)
        # para proteger su acceso y modificación directa desde fuera de la clase.
        # Su acceso se realiza a través de métodos públicos (`obtener_precio`
        # y `aplicar_descuento`).
        self.__precio = precio

    @abstractmethod
    def tipo_medicamento(self):
        # Este método es abstracto, lo que significa que cada subclase de Medicamento
        # debe proporcionar su propia implementación de cómo se determina su tipo.
        pass

    def obtener_precio(self):
        # Método público para acceder al precio encapsulado.
        return self.__precio

    def aplicar_descuento(self, porcentaje):
        # Método público para modificar el precio de forma controlada,
        # aplicando una lógica específica (solo si el porcentaje es válido).
        if 0 < porcentaje < 100:
            self.__precio -= self.__precio * (porcentaje / 100)


# 3. HERENCIA: 'Antibiotico' y 'Analgesico' son clases hijas que heredan
# de 'Medicamento'. Reutilizan la lógica común de 'Medicamento' (como la
# gestión del nombre, dosis y precio) y añaden sus propias características
# y comportamientos específicos.
class Antibiotico(Medicamento):
    def __init__(self, nombre, dosis_mg, precio, espectro):
        super().__init__(nombre, dosis_mg, precio) # Llama al constructor de la clase base
        self.espectro = espectro  # Característica específica del antibiótico

    # 4. POLIMORFISMO: Implementación específica del método abstracto 'tipo_medicamento'
    # para Antibiótico. Cada subclase implementa este método de una forma que le es propia,
    # pero todas responden al mismo nombre de método, demostrando polimorfismo.
    def tipo_medicamento(self):
        return "Antibiótico"


class Analgesico(Medicamento):
    def __init__(self, nombre, dosis_mg, precio, tipo_dolor):
        super().__init__(nombre, dosis_mg, precio) # Llama al constructor de la clase base
        self.tipo_dolor = tipo_dolor # Característica específica del analgésico

    # 4. POLIMORFISMO: Implementación específica del método abstracto 'tipo_medicamento'
    # para Analgésico.
    def tipo_medicamento(self):
        return "Analgésico"


# Clase Paciente: Una clase simple para representar a una persona.
class Paciente:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad


# Clase Receta médica: Demuestra cómo las clases de medicamentos interactúan.
class Receta:
    def __init__(self, paciente):
        self.paciente = paciente
        self.medicamentos = []

    def agregar_medicamento(self, medicamento):
        self.medicamentos.append(medicamento)
        print(f"{medicamento.nombre} agregado a la receta de {self.paciente.nombre}")

    def costo_total(self):
        # Aquí se ve el polimorfismo en acción indirectamente:
        # `m.obtener_precio()` se llama en diferentes tipos de objetos Medicamento,
        # y funciona uniformemente ya que todos implementan ese método.
        return sum(m.obtener_precio() for m in self.medicamentos)

    def mostrar_receta(self):
        print(f"\nReceta de {self.paciente.nombre}:")
        for m in self.medicamentos:
            # El polimorfismo permite llamar a `m.tipo_medicamento()`
            # sin saber si `m` es un Antibiotico o un Analgesico, y cada
            # objeto responderá con su tipo correcto.
            print(f"- {m.nombre} ({m.tipo_medicamento()}) - {m.dosis_mg} mg")
        print(f"Total a pagar: {self.costo_total():.2f}") # Formateo para precios


# USO DEL SISTEMA: Demostración de cómo interactúan las clases

# Crear instancias de medicamentos, utilizando las clases heredadas.
amoxicilina = Antibiotico("Amoxicilina", 500, 12000, "amplio")
ibuprofeno = Analgesico("Ibuprofeno", 400, 8000, "inflamatorio")

# Crear un paciente.
paciente1 = Paciente("Carlos", 30)

# Crear una receta para el paciente y agregar los medicamentos.
receta1 = Receta(paciente1)
receta1.agregar_medicamento(amoxicilina)
receta1.agregar_medicamento(ibuprofeno)

# Aplicar descuento a un medicamento, usando el método que controla
# el acceso al precio encapsulado.
ibuprofeno.aplicar_descuento(10) # Aplica un 10% de descuento al ibuprofeno

# Mostrar la receta final, que calcula el costo total y muestra los tipos de medicamento
# de forma polimórfica.
receta1.mostrar_receta()
```
