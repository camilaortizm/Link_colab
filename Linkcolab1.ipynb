```python
# Programacion orientada a objetos

# Creacion de un objeto
class Persona:
    def __init__(self, nombre, edad, altura):
        self.nombre = nombre
        self.edad = edad
        self.altura = altura

    def hablar(self):
        return f"Hola, mi nombre es {self.nombre}"

# Crear una instancia de ese objeto
persona1 = Persona ("Juan", 25, 1.75)
persona2 = Persona ("Maria", 30, 1.65)

# Uso de la clase
print(persona1.hablar())
print(persona2.hablar())
```

```python
from abc import ABC, abstractmethod

# Pilares de la POO

# 1. Abstracción
class CuentaBancaria(ABC):
    def __init__(self, titular, saldo_inicial):
        self.titular = titular
        self.__saldo = saldo_inicial  # encapsulado

    @abstractmethod
    def mostrar_tipo(self):
        pass

    # 2. Encapsulamiento
    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto
            print(f"Deposito de {monto} realizado. Nuevo saldo: {self.__saldo}")

    def obtener_saldo(self):
        return self.__saldo

    # 4. Polimorfismo
    def retirar(self, monto):
        if monto <= self.obtener_saldo():
            self.__saldo -= monto
            print(f"Retiro de {monto} realizado. Nuevo saldo: {self.__saldo}")
        else:
            print("Saldo insuficiente")


# 3. Herencia
class CuentaAhorro(CuentaBancaria):
    def __init__(self, titular, saldo_inicial, tasa_interes):
        super().__init__(titular, saldo_inicial)
        self.tasa_interes = tasa_interes

    def mostrar_tipo(self):
        return "Cuenta de Ahorro"


# Uso del código
mi_cuenta = CuentaAhorro("Juan", 1000, 0.05)

print(f"Tipo: {mi_cuenta.mostrar_tipo()}")
mi_cuenta.depositar(500)
print(f"Saldo: {mi_cuenta.obtener_saldo()}")
mi_cuenta.retirar(200)
print(f"Saldo: {mi_cuenta.obtener_saldo()}")
```
